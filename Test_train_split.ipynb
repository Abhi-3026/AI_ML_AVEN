**MULTI LINEAR REGRESSION**




*step1 : data collection*

```python
#companies code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

step 2:reading data

```python
df = pd.read_csv('/content/1000_Companies.csv')
df
```

STEP 3 : Assigning Xand y values

```python
X = df.drop('Profit',axis='columns')
y = df['Profit']
y
```

STEP 4 : Find correlation using heatmap

```python
sns.heatmap(df.corr())
```

STEP 5:Identify categorial data state and encode it

```python
dummies = pd.get_dummies(df['State'])
dummies
```

```python
merged = pd.concat([X,dummies],axis='columns')
merged
```

```python
X = merged.drop(['State','California'],axis = 'columns')
X
```

STEP 6 : Spliting of data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

STEP 7 :Train your model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

STEP 8: Prediction

```python
y_pred = model.predict(X_test)
y_pred
```

STEP 9 : Model metric calculation

```python
model.score(X_test,y_test)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/insurance_data.csv')
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker = '+', color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df['bought_insurance'],train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test, y_test)
```
